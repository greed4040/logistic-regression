```python
import numpy as np
import math

sgm = lambda x: 1/(1+np.exp(-x))
data1=[]
y_data=[]
for i in range(200):
    ar, y = np.random.randint(10, size=(3, 3)), 1    
    if i<=100: ar[2][0],ar[2][1],ar[2][2]=0,0,0;ar[1][0],ar[1][1],ar[1][2]=1,1,1; y=0
    data1.append(ar)
    y_data.append(y)
y = np.array([y_data])
train_set_x_orig=np.array(data1)
anorm=np.linalg.norm(y)
x = train_set_x_orig.reshape(train_set_x_orig.shape[0],-1).T/anorm
m = train_set_x_orig.shape[0]
print(train_set_x_orig.shape, m, y.shape, x.shape)
w,b=np.zeros((x.shape[0],1)),float(0)
learning_rate=0.05
```

```python
for i in range(100000):
    a=sgm(np.dot(w.T,x)+b)
    dz=np.subtract(a, y)/m
    dw=np.dot(x, dz.T)/m
    db=np.sum(a - y)/m
    w = w - dw * learning_rate
    b = b - db * learning_rate
    cost= -1/m*np.sum(y*np.log(a)+ (1-y)*np.log(1-a))  
    if i%1000==0: print("{0:5} {1:8.4f} {2:8.4f} cost:{3:8.4f}".format(i, np.sum(np.log(a)), np.sum((1-y)*np.log(1-a)), cost))
```
